```python
import os
import shutil

# Set base path
base_dir = "data/green_monkey/all_structure_files/"

# Define identifiers
temporal_keywords = ['temporal']

# Loop over chromosome folders
for chrom in os.listdir(base_dir):
    chrom_path = os.path.join(base_dir, chrom)
    if not os.path.isdir(chrom_path):
        continue

    # Create temporal_data and spatial_data folders if not exist
    temporal_dir = os.path.join(chrom_path, "temporal_data")
    spatial_dir = os.path.join(chrom_path, "spatial_data")
    os.makedirs(temporal_dir, exist_ok=True)
    os.makedirs(spatial_dir, exist_ok=True)

    # Move files to appropriate folder based on filename
    for fname in os.listdir(chrom_path):
        fpath = os.path.join(chrom_path, fname)
        if not os.path.isfile(fpath):
            continue

        if any(key in fname.lower() for key in temporal_keywords):
            shutil.move(fpath, os.path.join(temporal_dir, fname))
        else:
            shutil.move(fpath, os.path.join(spatial_dir, fname))

"File organization complete."
```

```python
import os
import shutil

# Set base path
base_dir = "data/green_monkey/all_structure_files/"

# Loop over chromosome folders
for chrom in os.listdir(base_dir):
    chrom_path = os.path.join(base_dir, chrom)
    spatial_path = os.path.join(chrom_path, "spatial_data")

    if not os.path.isdir(spatial_path):
        continue  # Skip if no spatial_data folder

    # Move all files inside spatial_data back to chromosome folder
    for fname in os.listdir(spatial_path):
        src = os.path.join(spatial_path, fname)
        dst = os.path.join(chrom_path, fname)

        # Move file
        shutil.move(src, dst)

    # # Optionally remove the empty spatial_data folder
    # os.rmdir(spatial_path)
```
