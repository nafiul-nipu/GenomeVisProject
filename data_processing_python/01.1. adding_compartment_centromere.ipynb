```python
import pandas as pd
import os
```

```python
before_path = 'data/MRC5/processed/MRC5/mock'
before_LIMS = '2501_147'
after_LIMS = '2501_149'
after_path = 'data/MRC5/processed/MRC5/229E'
time_hr = '48hr'
resolution = '250000'
structure_file_name = 'structure'
```

```python
centromere = pd.read_csv('data/MRC5/centromere_raw/Centromere_Assignments_Mapped_250000.csv')
# centromere.head()
compartment = pd.read_csv('data/MRC5/compartment_raw/MRC5_chromosome_compartment_scores_250000.csv')
```

```python
compartment.head()
```

```python
def add_initial_id_and_position(atom_file_path, hour, resolution, structure_file_name, lims):
    chr_folder_path = os.path.join(atom_file_path, hour, resolution)
    for chr_folder_name in os.listdir(chr_folder_path):

        # Check if the current item is a directory starting with 'chr'
        if os.path.isdir(chr_folder_path) and chr_folder_name.startswith('chr'):        
            # Navigate into the chr folder and look for structure-with-tracks.csv
            csv_file_path = os.path.join(chr_folder_path, chr_folder_name, f"{structure_file_name}.csv" )
            

            if os.path.exists(csv_file_path):
                df = pd.read_csv(csv_file_path)
                
                centromere_filtered = centromere[centromere['Chrom'] == chr_folder_name]
                centromere_list = centromere_filtered['Centromere'].tolist() 
#                 print(len(centromere_list))
                #removing the first entry as structure file starts from 1
                if(len(centromere_list) != len(df)):
                    centromere_list.pop(0)
#                 print(len(centromere_list))
                df['centromere'] = centromere_list
    
                
                com_lims = compartment[compartment['LIMS'] == lims]
                com_lims_filtered = com_lims[com_lims['Chrom'] == chr_folder_name]
                com_lims_list = com_lims_filtered['Compartment'].tolist()
                
                if(len(com_lims_list) != len(df)):
                    com_lims_list.pop(0)
                    
                df['compartment'] = com_lims_list
    
                
                
                
                # save_file_path = os.path.join(chr_folder_path, 'structure', '100kb', 'structure-with-id0.csv')
                save_file_path = os.path.join(chr_folder_path, chr_folder_name, 'structure.csv')
#                 print(save_file_path)
                df.to_csv(save_file_path, index=False)

                print(f"file saved to {save_file_path}")


add_initial_id_and_position(before_path, time_hr, resolution, structure_file_name, before_LIMS)
add_initial_id_and_position(after_path, time_hr, resolution, structure_file_name, after_LIMS)
```
