```python
import pandas as pd
import json
import numpy as np
import os
```

```python
folder_path = "data/extracted-node-link/cell7/with-gene/"
save_folder_path = "data/extracted-node-link/cell7/with-gene-csv/"

for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    if os.path.isfile(file_path) and file_name.endswith('.json'):
        print(file_name)
        with open(file_path, 'r') as file:
            data = json.load(file)
            df = pd.DataFrame(data['nodes'])
            # Create separate columns for the coord field
            df[['x', 'y', 'z']] = pd.DataFrame(df['coord'].tolist(), index=df.index)
#             df['id'] = df['id'] - 1
            df.drop(columns=['coord'], inplace=True)
            # Derive the CSV file name from the JSON file name
            csv_file_name = file_name.replace('.json', '.csv')
            csv_file_path = os.path.join(save_folder_path, csv_file_name)
            
            # Save the DataFrame as a CSV file
            df.to_csv(csv_file_path, index=False)
            print(f"DataFrame saved as {csv_file_path}")
```
