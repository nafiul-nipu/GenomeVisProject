```python
import networkx as nx
import json
import random
import numpy as np
```

```python
num_nodes = 100 
```

```python
# Generate a random graph
G = nx.fast_gnp_random_graph(n=num_nodes, p=0.02)

# Create a dictionary to store the graph data
graph_data = {
    'nodes': [],
    'edges': []
}

# Define the minimum and maximum values for each dimension
x_min, x_max = 0, 10
y_min, y_max = 0, 10
z_min, z_max = 0, 10

# Generate random positions for nodes within the specified range
node_positions = np.random.uniform(low=[x_min, y_min, z_min], high=[x_max, y_max, z_max], size=(G.number_of_nodes(), 3))


# Iterate over nodes and store their attributes
for node, data in G.nodes(data=True):
    node_data = {
        'id': node,
        'position': node_positions[node].tolist(),
    }
    graph_data['nodes'].append(node_data)

# Iterate over edges and store their attributes
for source, target, data in G.edges(data=True):
    source_position = graph_data['nodes'][source]['position']
    target_position = graph_data['nodes'][target]['position']
    
    edge_data = {
        'source': source,
        'target': target,
        'source_position': source_position,
        'target_position': target_position
    }
    graph_data['edges'].append(edge_data)

# Export the graph data to JSON file
with open('data/graph_{}.json'.format(num_nodes), 'w') as f:
    json.dump(graph_data, f, indent=4)

print("Graph data exported to graph_{}.json".format(num_nodes))

```
