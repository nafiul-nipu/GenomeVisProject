```python
edges = [
    {'source': 1, 'target': 2},
    {'source': 1, 'target': 3},
    {'source': 1, 'target': 1},
    {'source': 2, 'target': 2},
    {'source': 2, 'target': 3},
    {'source': 2, 'target': 1},
    {'source': 3, 'target': 14},
    {'source': 3, 'target': 2},
    {'source': 3, 'target': 3},
    {'source': 3, 'target': 4},
    {'source': 3, 'target': 1},
    {'source': 4, 'target': 11},
    {'source': 4, 'target': 5},
    {'source': 4, 'target': 3},
    {'source': 4, 'target': 4},
]
```

```python
'''expected output
[
[1,2,3,14], [1,3,4,11], [4,5]
]
'''
```

```python
paths = []
visited = set()
for edge in edges:
    source = edge['source']
    target = edge['target']
#     self visiting connections are excluded
    if source != target:
#         check if they are already visited or not
        if (source, target) not in visited and (target, source) not in visited:
            visited.add((source, target))
#             add the path
#             if the paths list empty add for the first onw
            if not paths:
                paths.append([source, target])
            else:
#                 iterate over all paths
                for path in paths:
#                   if source is present as the last element in any paths add it
                    if path[-1] == source:
                        path.append(target)
                        break
                else:
#                   if not found then create a new list
                    paths.append([source, target])
```
