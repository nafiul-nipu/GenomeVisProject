```python
import numpy as np
import pandas as pd
import json
```

```python
num_nodes = 100
```

```python
# file path
file_path = 'data/graph_clusters_{}.json'.format(num_nodes)
```

```python
# Open the JSON file and load its contents
with open(file_path, 'r') as file:
    data = json.load(file)
```

```python
data
```

```python
grouped_data = {}
for node in data['nodes']:
    kmeans_key = "cluster{}".format(node['Kmeans'])
    if kmeans_key not in grouped_data:
        grouped_data[kmeans_key] = {"nodes":[]}
    grouped_data[kmeans_key]['nodes'].append(node)

# Sort the grouped data based on the 'Kmeans' key
sorted_grouped_data = {k: grouped_data[k] for k in sorted(grouped_data.keys())}

sorted_grouped_data
```
